```python
import pickle
import pprint
import pickle
import json
from overcooked_ai_py.mdp.overcooked_mdp import OvercookedGridworld, OvercookedState
from overcooked_ai_py.planning.planners import MediumLevelActionManager

# Path to the .pkl file
file_path = "/Users/varshinichinta/Desktop/IRL/overcooked_ai/src/overcooked_demo/data/burger_test/New/HA/2025-04-13_19-23-30/result.pkl"
with open(file_path, "rb") as f:
    data = pickle.load(f) #is a trajectory

#pprint.pprint(data["trajectory"])
traj_len = len(data["trajectory"])# List of OvercookedState
mlam_params = {
    # MLAM planning options
    "wait_allowed": True,
    "counter_goals": ["onion", "tomato", "dish", "soup", "cheese", "burger", "bun"],
    "counter_drop": True,
    "counter_pickup": True,
    "num_goals": 1,
    "gamma": 0.99,
    "mdp_planner": "medium_level",
    "use_new_metrics": False,

    # JointMotionPlanner (low-level motion planning)
    "start_orientations": True,
    "same_motion_goals": False
}


# pprint.pprint(data["trajectory"][0]) 
# print(traj_len)
features_over_time = []
mdp = OvercookedGridworld.from_layout_name("burger_test")

print("mdp...")
mlam = MediumLevelActionManager.from_pickle_or_compute(mdp, mlam_params)

print(mlam)

i = 0
for time_step in data["trajectory"]:
    if i == 5:
        break
    state_obj = OvercookedState.from_dict(json.loads(time_step["state"]))
    features = mdp.featurize_state(state_obj, mlam)
    features_over_time.append(features)
    i += 1
print(features_over_time)
# pprint.pprint(data)
```
